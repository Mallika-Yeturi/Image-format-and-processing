# 5. Read bird and cat images, then implement blendImg function (10 points)

````python
# Function to blend two images
def blendImg(img1, img2, alpha=0.5):
    # Resize img1 to match img2's size
    img1_resized = cv2.resize(img1, (img2.shape[1], img2.shape[0]))
    # Blend the images using the alpha value
    blended_image = cv2.addWeighted(img1_resized, alpha, img2, 1 - alpha, 0)
    return blended_image

# Load the bird and cat images and convert them to grayscale
bird_img = cv2.imread('/Users/mallikayeturi/Desktop/Homework2/bird.png', cv2.IMREAD_GRAYSCALE)
cat_img = cv2.imread('/Users/mallikayeturi/Desktop/Homework2/cat.png', cv2.IMREAD_GRAYSCALE)

# Create figure for showing original and blended images in grayscale
plt.figure(figsize=(15, 8))

# Show the original bird image
plt.subplot(2, 3, 1)
plt.imshow(bird_img, cmap='gray')
plt.title('Original Bird Image')

# Show the original cat image
plt.subplot(2, 3, 2)
plt.imshow(cat_img, cmap='gray')
plt.title('Original Cat Image')

# Blend image with alpha = 0.3
blended_image1 = blendImg(bird_img, cat_img, alpha=0.3)
plt.subplot(2, 3, 3)
plt.imshow(blended_image1, cmap='gray')
plt.title('Blended Image (alpha=0.3)')

# Blend image with alpha = 0.5
blended_image2 = blendImg(bird_img, cat_img, alpha=0.5)
plt.subplot(2, 3, 4)
plt.imshow(blended_image2, cmap='gray')
plt.title('Blended Image (alpha=0.5)')

# Blend image with alpha = 0.9
blended_image3 = blendImg(bird_img, cat_img, alpha=0.9)
plt.subplot(2, 3, 5)
plt.imshow(blended_image3, cmap='gray')
plt.title('Blended Image (alpha=0.9)')

# Display all images
plt.tight_layout()
plt.show()
````

# 7. Read bridge image and add salt & pepper noise (10 points)

````python
# Function to add salt & pepper noise

def salt_pepper_noise(image, prob):
    noisy_image = np.copy(image)
    
    noise = np.random.rand(image.shape[0], image.shape[1])
    
    noisy_image[noise < prob] = 0
    
    noisy_image[noise > 1 - prob] = 255
    
    return noisy_image

# Step 1: Load the original bridge image in grayscale
bridge_image = cv2.imread('/Users/mallikayeturi/Desktop/Homework2/bridge.png', cv2.IMREAD_GRAYSCALE)

# Step 2: Add salt & pepper noise to the image (let's use 5% noise for this)
noisy_image = salt_pepper_noise(bridge_image, 0.05)

# Step 3: Now, let's create a plot to display both the original image and the noisy image
plt.figure(figsize=(10, 5))

# Show the original image 
plt.subplot(1, 2, 1)  
plt.imshow(bridge_image, cmap='gray')  
plt.title('Original Bridge Image')  

# Result
plt.subplot(1, 2, 2)  
plt.imshow(noisy_image, cmap='gray')  
plt.title('Salt & Pepper Noise Image')  

#Layout
plt.tight_layout()

# Display the images
plt.show()
````

# 8. Plot Gaussian function (10 points)

````python
# Function to compute the Gaussian distribution
def gaussian_function(x, mean, sigma):
    """This function computes the Gaussian distribution for a given x, mean, and sigma."""
    # Gaussian formula: the normal distribution equation
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / sigma) ** 2)

# Step 1: Generate x values
x_values = np.linspace(1, 10000, 1000)

# Step 2: Calculate Gaussian values
y_values = gaussian_function(x_values, mean=5000, sigma=1000)

# Step 3: Create the plot for the Gaussian curve
plt.figure(figsize=(8, 6)) 
plt.plot(x_values, y_values, color='b', label='Mean=5000, Sigma=1000')

# Step 4: Result
plt.title('Gaussian (Normal) Distribution', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Gaussian Distribution', fontsize=12)

# Step 5: Display a legend
plt.legend()

# Step 6: Show the plot
plt.show()
````

# 9. Add Random and Gaussian noise (20 points)

````python
# Function to add salt & pepper noise
def add_salt_pepper_noise(image, prob):
    noisy_image = np.copy(image)
    noise = np.random.rand(image.shape[0], image.shape[1])

    noisy_image[noise < prob] = 0  
    noisy_image[noise > 1 - prob] = 255  
    return noisy_image

# Function to add Gaussian noise
def add_gaussian_noise(image, mean=0, sigma=25):
    gaussian_noise = np.random.normal(mean, sigma, image.shape)
    noisy_image = np.clip(image + gaussian_noise, 0, 255) 
    return noisy_image.astype(np.uint8)

# Load an image 
image = cv2.imread('/Users/mallikayeturi/Desktop/Homework2/bridge.png', 0) 

# Add salt & pepper noise with 5% noise probability
salt_pepper_image = add_salt_pepper_noise(image, 0.05)

# Add Gaussian noise with mean 0 and standard deviation of 25
gaussian_image = add_gaussian_noise(image)

# Display the original, salt & pepper, and Gaussian noise images
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(salt_pepper_image, cmap='gray')
plt.title('Salt & Pepper Noise')

plt.subplot(1, 3, 3)
plt.imshow(gaussian_image, cmap='gray')
plt.title('Gaussian Noise')
plt.tight_layout()

plt.show()
````
